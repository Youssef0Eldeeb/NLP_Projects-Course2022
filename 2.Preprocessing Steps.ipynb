## Tokenization

```python
import nltk
tokenizedText = nltk.word_tokenize(myCorpus)
```

```python
#  print(tokenizedText)
```

## Stopword Removing

```python
from nltk.corpus import stopwords
stopWordList = stopwords.words('english')
```

```python
myCorpus = myCorpus.lower()
```

```python
stopwordCleanText = []
for token in tokenizedText:
    if token not in stopWordList:
        stopwordCleanText.append(token)
```

```python
# print(stopwordCleanText)
```

## Punctuations Removing

```python
# import string 
# puncList = string.punctuation
# for token in stopwordCleanText:
#     if token in puncList:
#         stopwordCleanText.remove(token)
```

```python
punc_stopwordCleanText =[]
import re
for token in stopwordCleanText:
    word = re.sub(r'[!"#\$%&\'()*\+\,-\./:;<=>?@[\\\]\^_`{|\}\~]+','',token)
    if word != '':
        punc_stopwordCleanText.append(word)
```

```python
print(punc_stopwordCleanText)
```

## Stemming

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
porter = PorterStemmer()
lancaster = LancasterStemmer()
```

```python
# stemming by lancater
stemLancasterText = []
for w in punc_stopwordCleanText:
    stemWord = lancaster.stem(w)
    stemLancasterText.append(stemWord)
```

```python
#print(stemLancasterText)
```

```python
# stemming by porter 
stemPorterText = []
for w in punc_stopwordCleanText:
    stemWord = porter.stem(w)
    stemPorterText.append(stemWord)
```

```python
#print(stemPorterText)
```

```python
# stemming by snowball
from nltk.stem.snowball import SnowballStemmer
englishStemmer = SnowballStemmer('english')

stemSnowballText = []
for w in punc_stopwordCleanText:
    stemWord = englishStemmer.stem(w)
    stemSnowballText.append(stemWord)
```

```python
#print(stemSnowballText)
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
word_Net_Leamatizer = WordNetLemmatizer()
```

```python
lemmaText = []
for w in punc_stopwordCleanText:
    stemWord = word_Net_Leamatizer.lemmatize(w)
    lemmaText.append(stemWord)
```
