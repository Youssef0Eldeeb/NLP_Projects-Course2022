```python
import nltk
myCorpus = nltk.data.load(r'./corpora/spam_corpus/spam.txt')
```

### Tokenization

```python
import nltk
tokenizedText = nltk.word_tokenize(myCorpus)
```

```python
for i in tokenizedText:
    if i == 'ham' or 'spam':
        tokenizedText.remove(i)
```

```python
#  print(tokenizedText)
```

### Punctuations Removing

```python
puncCleanText =[]
import re
for token in tokenizedText:
    word = re.sub(r'[!"#\$%&\'()*\+\,-\./:;<=>?@[\\\]\^_`{|\}\~]+','',token)
    if word != '':
        puncCleanText.append(word)
```

```python
# print(puncCleanText)
```

### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
word_Net_Leamatizer = WordNetLemmatizer()
```

```python
preText = []
for w in puncCleanText:
    stemWord = word_Net_Leamatizer.lemmatize(w)
    preText.append(stemWord)
```

```python
print(preText)
```

```python
import nltk
from nltk.util import ngrams
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
```

```python
def nGram(text,num):
    text = ' '.join(text)
    n_gram = ngrams(nltk.word_tokenize(text),num)
    return [' '.join(gram) for gram in n_gram]
nGram(preText,2)
```

```python
def Count(txt):
    fd = nltk.FreqDist(txt)
    return fd
```

```python
biGramCount = Count(nGram(preText,2))
uniGramCount = Count(nGram(preText,1))
# uniGramCount
biGramCount
```

```python
for k,v in biGramCount.items():
    t = k.split()
    countUnigram = uniGramCount[t[0]]
    p = v/countUnigram
    print(f'P({k}|{t[0]}) = {p}')
```
