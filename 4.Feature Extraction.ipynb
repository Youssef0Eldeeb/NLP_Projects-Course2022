```python
import nltk
myCorpus = nltk.data.load(r'./corpora/spam_corpus/spam.txt')
```

### Tokenization

```python
import nltk
tokenizedText = nltk.word_tokenize(myCorpus)
```

```python
for i in tokenizedText:
    if i == 'ham' or 'spam':
        tokenizedText.remove(i)
```

```python
#  print(tokenizedText)
```

### Punctuations Removing

```python
puncCleanText =[]
import re
for token in tokenizedText:
    word = re.sub(r'[!"#\$%&\'()*\+\,-\./:;<=>?@[\\\]\^_`{|\}\~]+','',token)
    if word != '':
        puncCleanText.append(word)
```

```python
# print(puncCleanText)
```

### Lemmatization

```python
from nltk.stem import WordNetLemmatizer
word_Net_Leamatizer = WordNetLemmatizer()
```

```python
preText = []
for w in puncCleanText:
    stemWord = word_Net_Leamatizer.lemmatize(w)
    preText.append(stemWord)
```

```python
print(preText)
```

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Binary Encoding
ve = CountVectorizer(binary=True)
ve.fit(preText)
pd.DataFrame(ve.transform(preText).toarray(),columns=sorted(ve.vocabulary_.keys()))
```

```python
#Counting
vec = CountVectorizer()
vec.fit(preText)
pd.DataFrame(vec.transform(preText).toarray(),columns=sorted(vec.vocabulary_.keys()))
```

```python
#TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
vecc = TfidfVectorizer()
vecc.fit(preText)
pd.DataFrame(vecc.transform(preText).toarray(),columns=sorted(vecc.vocabulary_.keys()))
```
